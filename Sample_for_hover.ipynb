```python
import pandas as pd
import numpy as np
import datetime
from bokeh.models import CategoricalColorMapper, NumeralTickFormatter, HoverTool
from bokeh.models import ColumnDataSource, Grid, LinearAxis, Plot, VBar
from bokeh.plotting import output_notebook, figure
from bokeh.io import reset_output, show, output_file


reset_output()
output_notebook()

data = {'Date':['2022-01-20', '2022-01-21', '2022-01-22'], 'DailyDoses':[734, 1121, 551]}

manual_sample = pd.DataFrame(data)

manual_sample['DateT']= pd.to_datetime(manual_sample['Date'])


sample_cds = ColumnDataSource(manual_sample)


sample_data = figure(title='Sample Vac', 
                     x_axis_type="datetime",
                     )

sample_data.vbar(x='DateT',
                 top='DailyDoses',
                 width=datetime.timedelta(days=0.5),
                 source=sample_cds)

sample_data.add_tools(HoverTool(tooltips=[
                                            ('Date', '@DateT{%F}'),
                                            ('Daily Doses','@{DailyDoses}'),
                                            
                                          ],
                     formatters={'@DateT' : 'datetime'},
                     mode='vline')
                               )

#hover.formatters = {'@Date' : 'datetime'}

show(sample_data)
```
